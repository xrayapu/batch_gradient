# gradient model 
## $ Batch\ gradient\ descent $

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as mp
import pandas as pd
```

## function
$$ f(x)= x^2 + x +1 $$
$$ g(x)=x^4 - 4x^2 +5 $$
$$ h(x)=x^5 - 2x^4 +2 $$

```python
def f(x):
    return x**2+x+1
def df(x):
    return 2*x+1
def g(x):
    return x**4-4*x**2+5
def dg(x):
    return 4*x**3 -8*x
def h(x):
    return x**5-2*x**4+2
def dh(x):
    return 5*x**4-8*x**3
```

```python
#make data
value=np.linspace(start=-3,stop=3,num=500)
value1=np.linspace(start=-2,stop=2, num=1000)
value2=np.linspace(start=-2.5,stop=2.5,num=1000)
```

```python
mp.figure(figsize=[15,5])

mp.subplot(1,2,1)
mp.title('f >>function',fontsize=23)
mp.ylim(0,8)
mp.plot(value,f(value))
mp.subplot(1,2,2)
mp.title('df',fontsize=23)
mp.plot(value,df(value))
mp.show()
```

```python
mp.figure(figsize=[15,5])
mp.subplot(1,2,1)
mp.plot(value1,g(value1))
mp.subplot(1,2,2)
mp.plot(value1,dg(value1))
mp.show()
```

```python
mp.figure(figsize=[15,5])
mp.subplot(1,2,1)
mp.plot(value2,h(value2))
mp.subplot(1,2,2)
mp.plot(value2,dh(value2))
mp.show()
```

# algorithm 

```python
def gradient(func, ini , rate, stop_point, loop):
    new =ini
    new_list=[new]
    slope_list=[func(new)]
    
    for n in range (loop):
        old=new
        slope=func(old)
        size =rate*slope
        new=old - size
        it= abs(new-old)
        new_list.append(new)
        slope_list.append(func(new))
        
        if(it< stop_point):
            break
        
    return new, new_list, slope_list
```

```python
local, local_min, slop=gradient(df,0.5,0.02,0.001,1000)
mp.figure(figsize=[15,5])
mp.subplot(1,2,1)
mp.plot(value1,f(value1))
mp.scatter(local_min,f(np.array(local_min)), color='#d62828' , alpha=0.4)
mp.subplot(1,2,2)
mp.plot(value1,df(value1))
mp.grid()
mp.scatter(local_min, slop,color='#d62828',alpha=0.4)
mp.show()
print('local min:', local)
print("cost function's cost:",dg(local))
print('number of runing loop',len(local_min))

```

```python
local, local_min, slop=gradient(dg,-2,0.01,0.0001,1000)
print('local min:', local)
print("cost function's cost:",dg(local))
print('number of runing loop',len(local_min))
mp.figure(figsize=[15,5])
mp.subplot(1,2,1)
mp.plot(value1,g(value1))
mp.scatter(local_min,g(np.array(local_min)), color='#d62828' , alpha=0.4)
mp.subplot(1,2,2)
mp.plot(value1,dg(value1))
mp.grid()
mp.scatter(local_min, slop,color='#d62828',alpha=0.4)
mp.show()
```

```python
local, local_min, slop=gradient(dh,2,0.01,0.001,500)
mp.figure(figsize=[15,5])
mp.subplot(1,2,1)
mp.xlim(-1.2,2.5)
mp.ylim(-1,4)
mp.plot(value2,h(value2))
mp.scatter(local_min,h(np.array(local_min)), color='#d62828' , alpha=0.4)
mp.subplot(1,2,2)
mp.plot(value2,dh(value2))
mp.grid()
mp.scatter(local_min, slop,color='#d62828',alpha=0.4)
mp.xlim(-1,2)
mp.ylim(-4,5)
mp.show()
print('local min:', local)
print("cost function's cost: ", h(local))
print('number of runing loop',len(local_min))
```
